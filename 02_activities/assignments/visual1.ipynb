```python
xls = pd.ExcelFile("mtcg-short-term-rentals-en-apr-25.xlsx")
df_month = xls.parse("Month over month comparison")
```

```python
region_data = df_month.iloc[3:20, [0, 13]]
region_data.columns = ['Region', 'Total_Listings_2025']
region_data = region_data.dropna()
region_data['Total_Listings_2025'] = pd.to_numeric(region_data['Total_Listings_2025'], errors='coerce')
region_data = region_data.dropna()
region_data = region_data.sort_values(by='Total_Listings_2025', ascending=False)
```

```python
plt.figure(figsize=(12, 6))
sns.set_theme(style="whitegrid")
sns.barplot(data=region_data, x='Total_Listings_2025', y='Region', palette='viridis')

plt.title("Short-Term Vacation Rental Statistics, April 2025", fontsize=16)
plt.xlabel("Total Listings")
plt.ylabel("Region")
plt.tight_layout()
plt.show()
```
